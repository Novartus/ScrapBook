```python
# New Html

html = """
<h1><a /><b> <th <td>
""" 
```

```python
soup = BeautifulSoup(html, "lxml") # Specifing lxml parser to avoid warning
print(soup.prettify())
```

```python
soup = BeautifulSoup(html, "html.parser") # Specifing html.parser parser to avoid warning
print(soup.prettify())
```

```python
soup = BeautifulSoup(html, "html5lib") # Specifing html5lib parser to avoid warning
print(soup.prettify())
```

```python
soup = BeautifulSoup(html, "xml") # Specifing xml parser to avoid warning
print(soup.prettify())
```

```python
soup = BeautifulSoup(html, "lxml-xml") # Specifing xml parser from lxml to avoid warning
print(soup.prettify())
```
